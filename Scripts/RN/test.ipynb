```python
def pdf_extract_total(url):
    # Request and Open PDF file
    try:
        rq = requests.get(url)
        pdf_data = pdfplumber.open(BytesIO(rq.content))
    except Exception as e:
        print(e)

    print(pdf_data.pages[0].extract_text())

    total_confirm = re.findall(
        r'(^\d{1,3}\.\d{3})[ ]\d|([ ]\d{1,3}\.\d{3}[ ])',
        pdf_data.pages[0].extract_text(),
        re.MULTILINE
    )
    total_death = re.findall(
        r'(^\d{1,3}\.\d{3})\n[ ]',
        pdf_data.pages[0].extract_text(),
        re.MULTILINE
    )
    print(total_confirm)
    print(total_death)
    pdf_data.close()

    return [int(total_confirm[0][0].replace(".", "")), int(total_death[0].replace(".", ""))]
```

```python
print(pdf_extract_total("http://www.adcon.rn.gov.br/ACERVO/sesap/DOC/DOC000000000246735.PDF"))
```
