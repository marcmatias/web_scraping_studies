```python

import csv
import camelot
import re

camelot.core.pd.set_option('display.max_rows', None)
camelot.core.pd.set_option('display.max_columns', None)
camelot.core.pd.set_option('display.width', None)
camelot.core.pd.set_option('display.max_colwidth', None)
```

```python
# Getting cities IBGE code
with open("lista_nomes_mun.csv", mode="r", encoding="utf-8") as fobj:
    reader = csv.reader(fobj)
    mun_name_list = [row[0].upper() for row in reader]
```

```python
def tables_appended(pages_positions):

    def pdf_tables(page, regions):
        table_result = camelot.read_pdf("boletim_covid_05_7_21pdf.pdf", pages=f"{page}", flavor="stream",
                                        split_text=True, table_areas=[regions])[0].df

        table_result = table_result.drop([0], axis=0).reset_index(drop=True)
        table_result.replace(
            to_replace='[ ]{0,}\\n[A-Za-z0-9]{0,}[ ]{0,}', value='', inplace=True, regex=True)
        table_result = table_result[table_result[0].map(
            lambda name: True if name in mun_name_list else False)]

        if(len(table_result.columns) > 2):
            table_result = table_result.drop(
                [0], axis=1).rename(columns={1: 0, 2: 1})
        return table_result

    for el in pages_positions:
        try:
            table = table.append(pdf_tables(
                el["page"], el["region"])).reset_index(drop=True)
        except:
            table = pdf_tables(el["page"], el["region"])
    return table
```
