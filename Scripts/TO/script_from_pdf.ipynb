```python
import camelot
```

```python
def pdf_tables():
    tables = camelot.read_pdf("boletim_covid_07_7_21pdf.pdf", pages="2-4", split_text=True, process_background=True)
    return tables
```

```python
tables = pdf_tables()
print(tables)

#20210520_TO.pdf
#20210630_TO.pdf

camelot.core.pd.set_option('display.max_rows', None)
camelot.core.pd.set_option('display.max_columns', None)
camelot.core.pd.set_option('display.width', None)
camelot.core.pd.set_option('display.max_colwidth', None)

for table in tables:
    table.df.replace(to_replace='[0-9]+[ ]{0,}\\n', value='', inplace=True, regex=True)
    display(table.df)
```
